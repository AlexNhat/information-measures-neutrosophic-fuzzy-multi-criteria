# Create NFSet

```python
def change(A):
    Aok = []
    for i in range(3):
        temp = []
        for j in range(4):
            if j == 0:
                temp.append(A[i][j]/10)
            if j == 1:
                temp.append(A[i][j]/10)
            if j == 2:
                temp.append(A[i][j+1]/10)
            if j == 3:
                temp.append(A[i][j-1]/10)
        Aok.append(temp)
    return(Aok)

matrix = np.array([change(A),change(A1),change(B),change(C),change(D)])
matrixx = np.array([change(AA),change(AA1),change(BB),change(CC),change(DD)])

def rtc(A):
    rtc = []
    for i in range(3):
        temp = []
        for j in range(5):
            temp.append(A[j][i])
        rtc.append(temp)
    return np.array(rtc)
matric = rtc(matrix)
matricc = rtc(matrixx)

mat = []
for i in range(3):
    mat.append(matric[i])
    mat.append(matricc[i])
nfs = NFSet(mat)
print(nfs.data)
```

```python
cost = []
rnf = RNF(nfs,cost)
print(rnf.data)
```

```python
for i in range(9):
    result = DecisionMaker(rnf = rnf, index = i)
    print("Similarity no.",i+1,":", result.best_alternative())
```
